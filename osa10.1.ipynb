```python
#10.5
class SuperSankari:
    def __init__(self, nimi: str, supervoimat: str):
        self.nimi = nimi
        self.supervoimat = supervoimat

    def __str__(self):
        return f'{self.nimi}, superkyvyt: {self.supervoimat}'

class SuperRyhma(SuperSankari):
    def __init__(self, nimi: str, kotipaikka: str):
        self._nimi = nimi
        self._kotipaikka = kotipaikka
        self._jasenet = []
        
    # Havainnointimetodit
    @property
    def nimi(self):
        return self._nimi
    @property
    def kotipaikka(self):
        return self._kotipaikka
    
    def lisaa_jasen(self, sankari: SuperSankari):
        self._jasenet.append(sankari)
        
    def tulosta_ryhma(self):
        print(f'{self.nimi},{self.kotipaikka}')
        print('Jäsenet:')
        for i in self._jasenet:
            print(i)
    
if __name__== "__main__":
    
    supermiekkonen = SuperSankari("Supermiekkonen", "Supernopeus, supervoimakkuus")
    nakymaton = SuperSankari("Näkymätön Makkonen", "Näkymättömyys")
    ryhma_z = SuperRyhma("Ryhmä Z", "Kälviä")

    ryhma_z.lisaa_jasen(supermiekkonen)
    ryhma_z.lisaa_jasen(nakymaton)
    
    ryhma_z.tulosta_ryhma()
```

```python
#10.6
# TEE RATKAISUSI TÄHÄN:

class Taikajuoma:
    def __init__(self, nimi: str):
        self._nimi = nimi
        self._ainekset = []

    def lisaa_aines(self, ainesosa: str, maara: float):
        self._ainekset.append((ainesosa, maara))

    def tulosta_resepti(self):
        print(self._nimi + ":")
        for aines in self._ainekset:
            print(f"{aines[0]} {aines[1]} grammaa")

class SalainenTaikajuoma(Taikajuoma):
    def __init__(self, nimi: str, salasana: str):
        super().__init__(nimi)
        self._salasana = salasana    
    
    def lisaa_aines(self, ainesosa: str, maara: float, salasana: str):
        if salasana == self._salasana:
            super().lisaa_aines(ainesosa, maara)
        else:
            raise ValueError('Väärä salasana')
            
    def tulosta_resepti(self, salasana: str):
        if salasana == self._salasana:
            super().tulosta_resepti()
        else:
            raise ValueError('Väärä salasana')

if __name__ == "__main__":

    kutistus = SalainenTaikajuoma("Kutistus maksimus", "hokkuspokkus")
    kutistus.lisaa_aines("Kärpässieni", 1.5, "hokkuspokkus")
    kutistus.lisaa_aines("Taikahiekka", 3.0, "hokkuspokkus")
    kutistus.lisaa_aines("Sammakonkutu", 4.0, "hokkuspokkus")
    kutistus.tulosta_resepti("hokkuspokkus")

```

```python
#10.7
class Raha:
    def __init__(self, eurot: int, sentit: int):
        self._eurot = eurot
        self._sentit = sentit

    def __str__(self):
        if self._sentit < 10:
            return (f"{self._eurot}.{self._sentit:02} eur")
        else:
            return f"{self._eurot}.{self._sentit} eur"
    
    def __eq__(self, toinen):
        return self._eurot == toinen._eurot and self._sentit == toinen._sentit
    
    def __lt__(self, toinen):
        s1 = str(self)
        s2 = str(toinen)
        return s1 < s2 
        
    def __gt__(self, toinen):
        s1 = str(self)
        s2 = str(toinen)
        return s1 > s2 
    
    def __ne__(self, toinen):
        s1 = str(self)
        s2 = str(toinen)
        return s1 != s2 
    
    def __add__(self, toinen):
        if self._sentit + toinen._sentit >= 100:
            yhteenlasku = Raha(1 ,0)
            yhteenlasku._eurot += self._eurot + toinen._eurot
            yhteenlasku._sentit = self._sentit + toinen._sentit - 100  
        else:
            yhteenlasku = Raha(0 ,0)
            yhteenlasku._sentit = self._sentit + toinen._sentit
            yhteenlasku._eurot = self._eurot + toinen._eurot
        
        return yhteenlasku
    
    def __sub__(self, toinen):
        if self._sentit - toinen._sentit < 0:
            vahennys = Raha(-1 ,0)
            vahennys._eurot += self._eurot - toinen._eurot
            vahennys._sentit = 100 + (self._sentit - toinen._sentit)  
        else:
            vahennys = Raha(0 ,0)
            vahennys._eurot = self._eurot - toinen._eurot
            vahennys._sentit = self._sentit - toinen._sentit
        
        if vahennys._eurot < 0:
            raise ValueError (f'negatiivinen tulos ei sallittu')
        
        return vahennys
    
if __name__ == '__main__':

    raha1 = Raha(4, 5)
    raha2 = Raha(4, 50)
    print(raha1)
    print(raha2)
    print(raha1 != raha2)
```

```python
#10.8
class Paivays:
    def __init__(self, pp:int,kk:int,vvvv:int):
        self._pp = pp
        self._kk = kk
        self._vvvv = vvvv
    
    def __repr__(self):
        return f'{self._pp}.{self._kk}.{self._vvvv}'
    
    #apufunktio kääntää pmvn
    def _pvm(self): 
        p = f'{self._vvvv}{self._kk:02d}{self._pp}'
        return p
    
    def __eq__(self, other):
        return self._pvm() == other._pvm()
    
    def __lt__(self, other):
        return self._pvm() < other._pvm()
        
    def __gt__(self, other):
        return self._pvm() > other._pvm()
    
    def __ne__(self, other):
        return self._pvm() != other._pvm()
        
    def pura_paivat(self):
        return self._vvvv * 360 + self._kk * 30 + self._pp
    
    def rakenna_paivat (self, n_paivat: int):
        v = n_paivat // 360
        jakojaannos = n_paivat % 360
        k = jakojaannos // 30
        p = jakojaannos % 30
        return Paivays(p, k, v)
    
    def __add__(self, lisa: int):
        uusi_paivat = self.pura_paivat() + lisa
        uusi_pvm = self.rakenna_paivat(uusi_paivat)
        return uusi_pvm
        
    def __sub__(self, other):
        fdays = self.pura_paivat() - other.pura_paivat()
        return abs(fdays)
        
if __name__ == "__main__":
    
    p1 = Paivays(4, 10, 2020)
    p2 = Paivays(2, 11, 2020)
    p3 = Paivays(28, 12, 1985)

    print(p2-p1)
    print(p1-p2)
    print(p1-p3)
```

```python
#10.8 ratkaisu
class Paivays:
    def __init__(self, pv: int, kk: int, vuosi: int):
        self.__pv = pv
        self.__kk = kk
        self.__vuosi = vuosi 
    def __str__(self):
        return f'{self.__pv}.{self.__kk}.{self.__vuosi}'
    # Vertailut on helppo tehdä, kun muutetaan päiväys päiviksi
    def __arvo(self):
        return self.__vuosi * 360 + self.__kk * 30 + self.__pv
    # Muutetaan päivien määrä takaisin päivämääräk
    def __pvm(self, paivat: int):
        kk = paivat // 30
        vuodet = kk // 12
        paivat -= kk * 30
        kk -= vuodet * 12
        return Paivays(paivat, kk, vuodet)
    def __lt__(self, toinen: "Paivays"):
        return self.__arvo() < toinen.__arvo()
    def __gt__(self, toinen: "Paivays"):
        return self.__arvo() > toinen.__arvo()
    def __eq__(self, toinen: "Paivays"):
        return self.__arvo() == toinen.__arvo()
    def __ne__(self, toinen: "Paivays"):
        return self.__arvo() != toinen.__arvo()
    def __add__(self, lisays_paivina: int):
        return self.__pvm(self.__arvo() + lisays_paivina)
    def __sub__(self, toinen: "Paivays"):
        # abs(x) antaa x:n itseisarvon
        return abs(self.__arvo() - toinen.__arvo())
    
```

```python
## 10.9
class Kauppalista:
    def __init__(self):
        self.tuotteet = []

    def tuotteita(self):
        return len(self.tuotteet)

    def lisaa(self, tuote: str, maara: int):
        self.tuotteet.append((tuote, maara))

    def tuote(self, n: int):
        return self.tuotteet[n-1][0]

    def maara(self, n: int):
        return self.tuotteet[n-1][1]
    
    def __iter__(self):
        self.n = 0
        return self
    
    def __next__(self):
        
        if self.n < self.tuotteita():
            t = self.tuotteet[self.n]
            self.n += 1
            return t
        
        else:
            raise StopIteration           
if __name__ == "__main__":
    
    lista = Kauppalista()
    lista.lisaa("banaanit", 10)
    lista.lisaa("omenat", 5)
    lista.lisaa("ananas", 1)
    
    for tuote in lista:
        print(f"{tuote[0]}: {tuote[1]} kpl")
```
