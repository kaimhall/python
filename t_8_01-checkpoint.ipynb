```python
#exec 8.4
class Kauppalista:
    def __init__(self):
        self.tuotteet = []
    def tuotteita(self):
        return len(self.tuotteet)
    def lisaa(self, tuote: str, maara: int):
        self.tuotteet.append((tuote, maara))
    def tuote(self, n: int):
        return self.tuotteet[n - 1][0]
    def maara(self, n: int):
        return self.tuotteet[n - 1][1]

def tuotteita_yhteensa(lista: Kauppalista):
    yhteensa = 0
    for i in range(lista.tuotteita()):
         yhteensa= yhteensa + lista.maara(i+1)
    return yhteensa

def main():  
    lista = Kauppalista()
    lista.lisaa("banaanit", 10)
    lista.lisaa("omenat", 5)
    lista.lisaa("ananas", 1)
    print(tuotteita_yhteensa(lista))
    
if __name__ == "__main__":
    main()    
```

```python
#exec8.9
from datetime import date

class Kirja:
    def __init__(self, nimi: str, kirjoittaja: str, genre: str, kirjoitusvuosi: date):
        self.nimi = nimi
        self.kirjoittaja = kirjoittaja
        self.genre = genre
        self.kirjoitusvuosi = kirjoitusvuosi

def genren_kirjat(kirjat: list, genre: str):
    rikos=[]
    for kirja in kirjat:
        if kirja.genre == genre:
            rikos.append(kirja)
    return rikos
        
def main():
    python = Kirja("Fluent Python", "Luciano Ramalho", "ohjelmointi", 2015)
    everest = Kirja("Huipulta huipulle", "Carina Räihä", "elämänkerta", 2010)
    norma = Kirja("Norma", "Sofi Oksanen", "rikos", 2015)

    kirjat = [python, everest, norma, Kirja("Lumiukko", "Jo Nesbø", "rikos", 2007)]

    print("rikoskirjoja ovat")
    for kirja in genren_kirjat(kirjat, "rikos"):
        print(f"{kirja.kirjoittaja}: {kirja.nimi}")
    
if __name__ =="__main__":
    main()
```

```python
# exec8.10

class VahenevaLaskuri:
    
    def __init__(self, arvo_alussa: int):
        self.arvo = arvo_alussa
        self.alkuarvo = arvo_alussa
    
    def tulosta_arvo(self):
        print("arvo:", self.arvo)
    
    def vahenna(self):
        if self.arvo <= 0:
            self.arvo == 0
        else:
            self.arvo -= 1
    
    def nollaa(self):
        self.arvo = 0
        
    def palauta_alkuperainen_arvo(self):
        self.arvo = self.alkuarvo

def main():
    laskuri = VahenevaLaskuri(50)
    laskuri.tulosta_arvo()
    laskuri.vahenna()
    laskuri.tulosta_arvo()
    laskuri.nollaa()
    laskuri.tulosta_arvo()
    laskuri.vahenna()
    laskuri.tulosta_arvo()
    laskuri.palauta_alkuperainen_arvo()
    laskuri.tulosta_arvo()
    
if __name__ == "__main__":
    main()
```

```python
#exec8.11
class  Lukutilasto:
    def __init__(self): 
        self.lukuja = 0
        self.sum = 0
        self.avg = 0
        self.luvut = []
        
    def lisaa_luku(self, luku:int):
        self.lukuja += 1
        self.sum += luku
        self.avg = self.sum / self.lukuja
        self.luvut.append(luku)
        
    def lukujen_maara(self):
        return self.lukuja
    
    def summa(self):
        return self.sum
        
    def keskiarvo(self):
        return self.avg
    
def main():
    olio1 = Lukutilasto()
    olio2 = Lukutilasto()
    olio3 = Lukutilasto()
    
    print('Anna lukuja:') 
    while True:
        n = int(input(""))
        if n == -1:
            break
        else:
            olio1.lisaa_luku(n)
            if n % 2 == 0:
                olio2.lisaa_luku(n)
            elif n % 2 != 0:
                olio3.lisaa_luku(n)
    
    print('Summa:',olio1.summa())
    print('Keskiarvo:',olio1.keskiarvo())
    print('Parillisten summa:',olio2.summa())
    print('Parittomien summa:',olio3.summa())
    
if __name__ == "__main__":
    main()
```

```python
#exec8.11 - ratkaisu
class Sekuntikello:
    def __init__(self):
        self.sekunnit = 0
        self.minuutit = 0
    
    def __str__(self):
        return f'{self.minuutit:02}:{self.sekunnit:02}'    
    
    def tick(self):
        self.sekunnit += 1
        if self.sekunnit == 60:
            self.sekunnit = 0
            self.minuutit += 1
            if self.minuutit == 60:
                self.minuutit = 0
            
def main():
    kello = Sekuntikello()
    for i in range(3600):
        print(kello)
        kello.tick()
    
if __name__ == "__main__":
    main()
```

```python
#exec 8.12
class Maksukortti:
    def __init__(self, alkusaldo: float):
        self.saldo = alkusaldo
    
    def __str__(self):
        return f'Kortilla on rahaa {self.saldo:.1f} euroa'
    
    def syo_edullisesti(self):
        if self.saldo < 2.60:
            self.saldo = self.saldo
        else:
            self.saldo -= 2.60
            
    def syo_maukkaasti(self):
        if self.saldo < 4.60:
            self.saldo = self.saldo
        else:
            self.saldo -= 4.60
    
    def lataa_rahaa(self, maara:int):
        if maara < 0:
            raise ValueError('Kortille ei saa ladata negatiivista summaa')
        else:
            self.saldo += maara 
        

pekan_kortti = Maksukortti(20)
matin_kortti = Maksukortti(30)
pekan_kortti.syo_maukkaasti()
matin_kortti.syo_edullisesti()
print('Pekka:',pekan_kortti)
print('Matti:',matin_kortti)
pekan_kortti.lataa_rahaa(20)
matin_kortti.syo_maukkaasti()
print('Pekka:',pekan_kortti)
print('Matti:',matin_kortti)
pekan_kortti.syo_edullisesti()
pekan_kortti.syo_edullisesti()
matin_kortti.lataa_rahaa(50)
print('Pekka:',pekan_kortti)
print('Matti:',matin_kortti)
```

```python
#exec8.13
class Sarja:
    def __init__(self, nimi:str, kaudet:int,genre: list):
        self.nimi = nimi
        self.kaudet = kaudet
        self.genre = genre
        self.arvostelut = [0]
        self.lukumaara = 0
        self.keskiarvo = 0
         
        
    def __str__(self):
        genrelista = ", ".join(self.genre)
        if len(self.arvostelut) == 1:
            arv_teksti = 'ei arvosteluja'
        else:
            arv_teksti = 'arvosteluja {}, keskiarvo {:.1f} pistettä'.format(self.lukumaara, self.keskiarvo)
            
        return ('{} ({} esityskautta)\ngenret: {}\n{}'\
                .format(self.nimi,self.kaudet,genrelista,arv_teksti))
    
    def arvostele(self, arvosana: int):
        if arvosana >= 0 and arvosana <= 5:
            self.arvostelut.append(arvosana)
            self.lukumaara += 1
            self.keskiarvo = sum(self.arvostelut)/self.lukumaara
            
def arvosana_vahintaan(arvosana: float, sarjat: list):
    sarja = []
    for i in sarjat:
        if i.keskiarvo >= arvosana:
            sarja.append(i)
    return sarja

def sisaltaa_genren(genre: str, sarjat: list):
    sarja = []
    for i in sarjat:
        if (genre in i.genre):
            sarja.append(i)
    return sarja
    
def main():
    
    s1 = Sarja("Dexter", 8, ["Crime", "Drama", "Mystery", "Thriller"])
    s1.arvostele(5)

    s2 = Sarja("South Park", 24, ["Animation", "Comedy"])
    s2.arvostele(3)

    s3 = Sarja("Friends", 10, ["Romance", "Comedy"])
    s3.arvostele(2)

    sarjat = [s1, s2, s3]
 
    print("arvosana vähintään 4.5:")
    for sarja in arvosana_vahintaan(4.5, sarjat):
        print(sarja.nimi)
    
    print("genre Comedy:")
    for sarja in sisaltaa_genren("Comedy", sarjat):
        print(sarja.nimi)
        
if __name__ == "__main__":
    main()
```
